```python
import pandas as pd
a={
       "Id": [1, 2, 3, 4, 5, 6],
       "Name": ["Pradhan", "Venu", "Madhurima", "Nireekshan",
        "Shafi", "Veeru"],
       "Subject": ["English", "Java", "Html", "Python", "C", "dotnet"]
       }
b =   {
        "Id": [11, 12, 13, 14, 15, 16],
        "Name": ["Srinu", "Sumanth", "Neelima", "Daniel", "Arjun",
        "Veeru"],
        "Subject": ["Java", "Html", "Cpp", "Python", "C", "dotnet"]
        }
df1=pd.DataFrame(a)
df2=pd.DataFrame(b)
print(df1)
print(df2)
inner_merge=pd.merge(df1,df2,on="Subject",how="inner")
print(inner_merge)
left_merge=pd.merge(df1,df2,on="Subject",how="left")
print(left_merge)
right_merge=pd.merge(df1,df2,on="Subject",how="right")
print(right_merge)
outer_merge=pd.merge(df1,df2,on="Subject",how="outer")
print("outer_merge")
```
